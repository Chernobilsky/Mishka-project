```python
#Если же мы хотим матрицу со случайнымим знгачениями, то делаем np.random.rand
matrix_rand = np.random.rand(3, 3)
print('А это матрица со случайными значениями:')
print(matrix_rand)
```

```python
#если хотим значения ячеек не от 0 до 1, а от -10 до 20 (например), то немного модернизируем:
print('Тут значения матрицы целые числа от -10 до 20:')
print(np.random.randint(-10, 21, size=(3, 6)))
print()

print('А тут значения матрицы действительные числа от -10 до 20:')
print(np.random.rand(3, 6)*30-10)
print()
```

```python
# Изх обычной матрицы мы можем выделить диагональную с помощью np.diag:
# создаем сначала матрицу со случайным содержимым ячеек:
m_test = np.random.randint(-10, 21, size=(3, 6))
print('Вот исходная матрица')
print(m_test)
print()
print('А вот вектор из ячеек на диагонали исходной матрицы:')
# А дальше вычленяем из нее диагональ:
# np.diag делаем вектор яз ячеек, находящихся на диагонали исходной матрицы 
print(np.diag(m_test))
print()
print('А вот диагональная матрица, получившаяся из исходной матрицы')
print(np.diag(np.diag(m_test)))
```

```python
# а еще можно создавать матрицы с помощью битблиотеки Sympy, но это специфические символьные вычисления
# О различиях между SymPy и NumPy https://chat.openai.com/share/4790a2bb-b6aa-45eb-83e1-afbaa9ef549a 

from sympy import *
diag(-1, ones(2,2), Matrix([5,7,5]))
```

**Итоги:**
- мы создаем векторы с помощью `np.array`, нулевые векторы длины k `np.zeros((k,))`, единичные векторы длины k - `np.ones((1, k))`
- мы можем создавать матрицы с помощью `np.array`.  Можем и с помощью `np.matrix`, но `np.array` предпочтительнее, поэтому будем разбираться с ним, а `np.matrix` использовать только если в `np.array` нет нужных возможностей. 
- есть готовые инструменты создания нулевых матриц с помощью `np.zeros`, и с единичной диагональю `np.eye`. Единичная матрица размера k создается `np.eye(k)`, а вот такая `np.eye(3, 6, 1)` создает смещенную диагональную матрицу. 
- матрицы заданной формы со случайным содержанием ячеек создаются вот так `np.random.rand(3, 6)*30-10` (если нам нужно с дробными числами) или вот так `np.random.randint(-10, 21, size=(3, 6))` (если нужны целые числа)
- есть отдельная библиотека `sympy`, где тоже есть матрицы. Но [это специфические символьные вычисления](https://chat.openai.com/share/4790a2bb-b6aa-45eb-83e1-afbaa9ef549a), не уверена, что они могут быть полезны при обработке изображений

# Транспонирование матрицы

```python
test_for_transpose = np.random.randint(-5, 15, size=(3, 5))
print('Исходная матрица, которую мы будем транспонировать:')
print(test_for_transpose)

print()
print('А теперь ее транспонируем, получилось вот так:')
print(test_for_transpose.transpose())
# тут тип объекта тот же - np.array, двумерный массив
```

```python
# есть и другой вариант транспонирования матрицы 
# у нас получается объект библиотеки symphy, то есть это функция из библиотеки sympy
print('класс объекта в результате использования функции transpose:', type(transpose(test_for_transpose)))
transpose(test_for_transpose)
```

```python
test_for_transpose.transpose()
```

```python
from colorama import init, Fore, Style
# В случае двумерной матрицы, где не нужно указывать порядок осейй, вместо метода matrix.transpose() можно писать matrix.T
check_transpose = np.random.randint(-10, 50, size=(3, 5))
print(Style.BRIGHT +'равны ли матрицы, транспонированные методом .transpose() и методом .T?', 
      np.array_equal(check_transpose.T, check_transpose.transpose()))
```

Действуют общие правила для транспонированных матриц, которые показывать в коде не надо, и итак понятно:
- (A<sup>T</sup>)<sup>T</sup> =A
- (A+B)<sup>T</sup> =  A<sup>T</sup> + B<sup>T</sup>
- (AB)<sup>T</sup> = B<sup>T</sup>A<sup>T</sup>

# Действия над матрицами. Сложение, умножение матрицы на число, умножение матрицы на матрицу

```python
# Сложение матриц очевидно, можно не показывать
# Умножение матрицы на число задается напрямую и работает очень просто
test_for_mult = np.random.randint(-10, 10, size=(3, 5))
print('Исходная матрица:')
print(test_for_mult)

print()
print('А теперь та же матрица, умноженная на два:')
print(2*test_for_mult)
```

**А теперь умножение матриц:**

```python
A_for_mult = np.random.randint(-10, 10, size=(3, 5))
print('Это матрица А:')
print(A_for_mult)
print()
B_for_mult = np.random.randint(-10, 10, size=(5, 3))
print('А это матрица В:')
print(B_for_mult)
print()
print('А это произведение матриц А и В:')
print(A_for_mult.dot(B_for_mult))
print()
print('А это произведение матриц B и A:')
print(B_for_mult.dot(A_for_mult))

print()
print('А это произведение B транспонированной на A транспонированную:')
print(B_for_mult.T.dot(A_for_mult.T))
print('Которое, как мы видим, равно произведению А на В')
```

```python
A_for_mult_E = np.random.randint(-5, 5, size=(3, 5))
print('Это матрица А размера 3х5:')
print(A_for_mult_E)
print()

print('Умножим матрицу А на единичную 5х5')
print(A_for_mult_E.dot(np.eye(5)))
print()


print('А теперь умножим единичную матрицу 3х3 на матрицу А')
print(np.eye(3).dot(A_for_mult_E))
print()

print('А если мы хотим, чтобы ячейки получившейся матрицы имели формат целого числа,')
print('меняем тип ячеек для единичной матрицы')
print(np.eye(3, dtype=int).dot(A_for_mult_E))
```

# Детерминант матрицы

```python
# Для детерминанта существует готовая функция np.linalg.det(A)
print('Марица А:')
A = np.random.randint(-5, 10, size=(5, 5))
print(A)
print()
print('Детерминант матрицы А равен ', round(np.linalg.det(A)))
```

```python
# детерминант транспонированной матрицы равен детерминанту исходной 
A = np.random.randint(-5, 10, size=(4, 4))
print('детерминант матрицы А',round(np.linalg.det(A)))
print('детерминант транспонированной матрицы',round(np.linalg.det(A.T)))
```

```python
# Если у матрицы две строки будут равные, то ее дискриминант равен нулю 
# сделаем матрицу с двумя равными строками

A = np.random.randint(-5, 10, size=(4, 4))
A[2, :] = A[1, :]
print('Вот матрица А с двумя равными строчками')
print (A)

print('Детерминант матрицы А равен', round(np.linalg.det(A)))
```

```python
print('Марица А:')
A = np.random.randint(-5, 10, size=(5, 5))
print(A)
print()
print('Детерминант матрицы А равен ', round(np.linalg.det(A)))
k = 3

print()
print('Теперь умножим одну строчку матрицы А в', k, 'раз')

i=2
B=np.copy(A)
B[i, :] *= k

print()
print('Теперь марица А равна:')
print(B)
print('И ее дискриминант вырос в', round(np.linalg.det(B)/np.linalg.det(A)),'раз')
print()

print('А если мы всю матрицу А умножим на',k,', то дискриминант вырастет в',round(np.linalg.det(k*A)/np.linalg.det(A)), 'раз' )
print('То есть в', k ,'в', A.shape[0], 'степени раз'  )

# и со столбцами то же самое. 
```

**Следствие из свойств 4 и 5:** 

Если одна из строк матрицы равна произведению другой на число ***k***, то детерминант будет равен ***k умножить на 0 = 0***

```python
A = np.random.randint(-3, 8, size=(5, 5))
print('это матрица А')
print(A)
print()
print('Это вектор b')
b = np.random.randint(low = -4, high = 5, size = 5)
print(b)
print()

B = np.copy(A)
i=1
B[i,:] = b
print('Это матрица В. Тут', i+1, 'строчка равна вектору в, остальные строки совпадают со строками матрицы А')
print(B)
print()

print('Теперь пусть будет матрица С. У нее все строки, кроме', i+1, 'совпадают со строками матриц А и В, а', 
      i+1, 'строка сумма строк матриц А и В')
C = np.copy(A)
C[i] +=b
print(C)
print()
print("В результате:")
print('детерминант А равен', round(np.linalg.det(A)))
print('детерминант B равен', round(np.linalg.det(B)))
print('детерминант C равен', round(np.linalg.det(C)))
```

**А следствие из свойств 4,5 и 6 это свойства 7 и 8:**

**Важное про сложение строк матрицы с векторами:**
- `A[i,:] += b` означает то же самое, что и `C[i:,] +=b`. В этом случае к  i+1 строке матрицы С добавляется вектор b.
- А вот `A[i:,] +=b` означает совсем другое. В этом случае вектор b прибавляется к каждой строке матрицы A, начиная с i-й строки и до конца матрицы. И это дает совсем другой результат.
